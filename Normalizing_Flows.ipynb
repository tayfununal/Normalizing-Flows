# Normalizing Flows


---



In simple words, normalizing flows is a series of simple functions which are invertible, or the analytical inverse of the function can be calculated. For example,

\begin{align}
                    f(x) = x+1
\end{align}

is a reversible function because for each input, a unique output exists and vice-versa whereas

\begin{align}
                    f(x) = x^2
\end{align}
is not a reversible function. Such functions are also known as bijective functions.
